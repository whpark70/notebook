```python
import matplotlib as mpl
import matplotlib.pylab as plt
%matplotlib inline

from matplotlib import rc
import matplotlib.font_manager as fm

font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False

import numpy as np
```

```python
plt.plot([10, 20,30,40], [1,4,9,16], c='b', lw=5, ls='--', marker='o', ms=15, mec='g',mew=5, mfc='r')
plt.title("스타일 적용 예")
plt.show()
```

```python
# 그림 범위 지정
plt.title("x 축, y축의 범위 설정")
plt.plot([10,20, 30, 40], [1,4, 9, 16],
        c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(0,50)
plt.ylim(-10, 30)
plt.show()
```

```python
# 틱 설정
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("x축과 y축의 tick label 설정")
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi/2, 0,np.pi/2, np.pi])
plt.yticks([-1,0, 1])
plt.show()
```

```python
# tick lable에는 $$사이에 LaTex 수학 문자식을 넣을 수 있다.
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("LaTex, 문자열로 tick lable 정의")
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],['Low', "Zero", 'High'])
plt.show()
```

```python
# grid
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.title("Grid 추가")
plt.plot(X,C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,1],['Low', "Zero", 'High'])
plt.grid(True)
plt.show()
```

```python
# 여러 개의 선 그리기
t = np.arange(0.,5.,0.2)
plt.title("라인 플롯에서 여러 개의 선 그리기")
plt.plot(t,t,'r--', t, 0.5*t**2,'bs:',t, 0.2*t**3, 'g^-' )
plt.show()
```

```python
plt.title("복수의 plot 명령을 한 그림에서 표현")
plt.plot([1,4,9,16], c='b', lw=5, ls='--', marker='o', ms=15, mec='g', mew=5, mfc='r')
plt.plot([9,16,4,1], c='k', lw=3, ls=':', marker='s', ms=10, mec='m', mew=5, mfc='c')
plt.show()
```

```python
# legend
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X,C, ls='--', label='cosine')
plt.plot(X,S, ls=':', label='sine')
plt.legend(loc=2)
plt.show()
```

```python
# x축, y축 타이틀
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X,C, label='cosine')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title("Cosine plot")
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2,1,1)
plt.plot(x1,y1, 'yo')
plt.title("A table of 2 subpolts")
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(x2,y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
# subplots 명령으로 복수의 Axes객체를 동시에 생성할 수도 있다. 
# 이때는 2차원 ndarray형태로 객체가 반환된다.
fig, axes = plt.subplots(2,2)

np.random.seed(0)
axes[0,0].plot(np.random.rand(5))
axes[0,0].set_title("axes 1")
axes[0,1].plot(np.random.rand(5))
axes[0,1].set_title("axes 2")

axes[1,0].plot(np.random.rand(5))
axes[1,0].set_title("axes 3")
axes[1,1].plot(np.random.rand(5))
axes[1,1].set_title("axes 4")
plt.tight_layout()
plt.show()
```

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax0.set_xlabel("공유되는 축")
plt.show()
```
